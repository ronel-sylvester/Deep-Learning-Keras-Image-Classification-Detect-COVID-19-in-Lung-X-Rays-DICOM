### Checking out 1 Mask Image within .nii Image

We use nibabel library to convert DICOM .nii Images to multidimensional arrays of size (301, 512, 512). This basically means that there are 301 slices of lung mask images for each CT Scan of a coronavirus patient.

```python
CT = nibabel.load('C:/Users/ronel/OneDrive/Documents/Projects/COVID19_DICOM/ct_scans/coronacases_org_001.nii')
CT_array = CT.get_fdata()
CT_array.shape
CT_array = CT_array.T
```

### Displaying Range of Mask Images

```python
lung_array = list(range(0, 301, 50))

plt.figure()

f, axarr = plt.subplots(len(lung_array), 1) 
f.set_figheight(30)
f.set_figwidth(30)

for i in range(len(lung_array)):
    axarr[i].imshow(CT_array[lung_array[i]])
    axarr[i].set_title(f'Mask {lung_array[i]}')
```
